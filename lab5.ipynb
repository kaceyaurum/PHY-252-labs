### 4

Fit your data to a Gaussian distribution with mean $\mu$ and variance $\sigma^2$, namely $n(\mu , \sigma ^2)\sim \frac{A}{\sqrt{2\pi \sigma ^2}}e(\frac{-(x-\mu)^2}{2\sigma ^2})$ and report $𝐴 \pm \delta𝐴$, $\mu \pm \delta\mu$ and $\sigma \pm \delta\sigma$ with units.

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot(111)
ax.hist(ourmacrostates)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))

ourhist, bins = np.histogram(ourmacrostates)
mids = [(bins[i]+bins[i-1])/2 for i in range(1, len(bins))]

mu = np.average(ourmacrostates)
std = np.std(ourmacrostates)

xmin, xmax = 0, 19
xs = np.arange(xmin, xmax, 1)
A = 1
N = 18
p = [A/np.sqrt(2*np.pi*np.sqrt(n*(N-n)*N))*np.e**(-(x-mu)**2/(2*np.sqrt(n*(N-n)*N)**2)) for x in xs]
ax.plot(xs, p)
ax.set_xlabel('macrostate (# up)')
ax.set_ylabel('frequency (#)')
ax.set_title('Fit results: mu = '+str(mu)+', std = '+str(std))
```

What do these quantities represent?

$A$ is the scaling factor. $\delta A$ is the standard deviation of the scaling factor.

$\mu$ is the average number of eyes up, so it makes sense that it would be around the middle of the range of possible values. $\delta\mu$ is the standard deviation of $\mu$, which is a measure of how much the data deviates from its average.

$\sigma$ is also the standard deviation of $\mu$. $\delta \sigma$ is the standard deviation of the standard deviation of $\mu$, which is a measure of the variation in the spread of the observed macrostates.
