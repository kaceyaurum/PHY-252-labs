```python
import pandas as pd
import numpy as np
from pylab import polyfit
import scipy.optimize
from matplotlib.ticker import MaxNLocator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('gain1.csv')
data = data.drop([0,1])
data = data.reset_index()
```

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot(111)
ax.scatter(data['Column1'], data['Column2'])
ax.set_title('Intensity vs Angular Position (GAIN=1)')
ax.set_xlabel(r'Angular Position ($^\circ$)')
ax.set_ylabel('Intensity (% of maximum)')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.figure.savefig('gain1.png')
```
